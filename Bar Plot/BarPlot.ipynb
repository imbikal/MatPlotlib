# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.style.use('dark_background')
```

```python
subjects = ['Math', 'Science', 'English', 'Nepali', 'Computer']
marks = [85, 90, 75, 80, 95]
```

## Bar plot 

```python
plt.bar(subjects,marks)
plt.show()
```

```python
plt.bar(subjects,marks,color = 'green')
plt.show()
```

## Different colors for different building/bars

```python
colors = ['red','blue','green','orange','purple']
plt.bar(subjects,marks, color = colors)
plt.show()
```

## Width of buildings/bars

```python
plt.bar(subjects,marks, color = colors, width = 0.5)
plt.show()
```

## Provide border to the buildings/bars

```python
plt.bar(subjects,marks, color = colors, edgecolor = 'white', width = 0.8)
plt.show()
```

## Border line width

```python
plt.bar(subjects,marks, color = colors, edgecolor = 'white', width = 0.8,linewidth = 4)
plt.show()
```

## Border line style

```python
plt.bar(subjects,marks, color = colors, edgecolor = 'white', width = 0.8,linewidth = 4, linestyle = '--')
plt.show()
```

## Plotting bar in horizontal direction

```python
plt.barh(subjects,marks, color = colors)
plt.show()
```

## Two plots in the same graph

```python
subjects = ['Maths', 'English', 'Science', 'Social Studies', 'Computer']
marks1 = [89, 90, 45, 78, 99]
marks2 = [78, 56, 34, 90, 12]
```

```python
plt.figure(figsize = (8,8))

plt.bar(subjects, marks1)
plt.bar(subjects, marks2)

plt.xlabel('Subjects')
plt.ylabel('Marks')

plt.show()
```

## Compairing side by side

```python
subjects_len = np.arange(len(subjects))
width = 0.4     
```

```python

plt.figure(figsize = (8,8))

plt.bar(subjects_len, marks1, width = width)
plt.bar(subjects_len + width, marks2, width = width)

plt.xlabel('Subjects')
plt.ylabel('Marks')

plt.show()
```

## Compairing with Colors

```python
plt.figure(figsize = (8,8))

plt.bar(subjects_len, marks1, width = width, color = colors)
plt.bar(subjects_len + width, marks2, width = width, color = colors, alpha = 0.5)

plt.xlabel('Subjects')
plt.ylabel('Marks')

plt.show()     
```
