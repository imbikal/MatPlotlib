# Pie Plot

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd     
```

```python
plt.style.use('dark_background')
```

```python
classes = ['Physics', 'Chemistry', 'Maths', 'English', 'Computer']
marks = [89, 45, 78, 82, 90]     
```

```python
plt.pie(marks, labels = classes)
plt.show()
```

## Custom colors

```python
colors = ['red', 'blue', 'green', '#9803fc', '#03c2fc']

plt.pie(marks, labels = classes, colors=colors)
plt.show()     
```

## Percentage for each slices

```python
plt.pie(marks, labels = classes, colors=colors, autopct = '%0.1f%%')
plt.show()
```

## Explode values (Popping up slices)

```python
explode_values = [0.1, 0, 0, 0, 0]

plt.pie(marks, labels = classes, colors=colors, autopct = '%0.1f%%', explode=explode_values)
plt.show()
```

## Shadow in popped up slice

```python
plt.pie(marks, labels = classes, colors=colors, autopct = '%0.1f%%', 
                                            explode=explode_values, 
                                            shadow = True)
plt.show()
```

## Changing the size of pie-chart

```python
plt.pie(marks, labels = classes, colors=colors, autopct = '%0.1f%%', 
                                            explode=explode_values, 
                                            radius = 1.6)
plt.show()
```

## Text Properties

```python
textprops = {'fontsize':14, 'color':'k'}

plt.pie(marks, labels = classes, colors=colors, autopct = '%0.1f%%', 
                                            explode=explode_values, 
                                            radius = 1.6, textprops = textprops)
plt.show()
```

## Wedge Properties (Properties of slices)

```python
textprops = {'fontsize':14, 'color':'k'}
wedgeprops = {'linewidth':3, 'linestyle':'--', 'edgecolor':'white'}

plt.pie(marks, labels = classes, colors=colors, autopct = '%0.1f%%', 
                                            explode=explode_values, 
                                            radius = 1.6, textprops = textprops, 
                                            wedgeprops = wedgeprops)
plt.show()
```

## Legend to denote slices

```python
plt.figure(figsize = (6,6))

plt.pie(marks, labels = classes, colors=colors, autopct = '%0.1f%%', 
                                            explode=explode_values, 
                                            textprops = textprops)

plt.title('Subjects and Average Scores')
plt.legend()
plt.show()
```
