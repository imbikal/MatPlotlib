# Img Show Plot

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.image as mpimg
```

## Reading Image using Matplotlib

```python
img = mpimg.imread('lemon.jpg')
```

```python
plt.imshow(img)
plt.show()
```

## Reading Image using OpenCV

```python
import cv2
```

```python
imgcv2 = cv2.imread('lemon.jpg')
```

```python
plt.imshow(imgcv2)
plt.show()
```

```python
imgcv2 = cv2.cvtColor(imgcv2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imgcv2)
plt.show()
```

## Changing Aspect

```python
plt.imshow(img, aspect = 0.5)
plt.show()
```

```python
img1 = mpimg.imread('lemon.jpg')
img2 = mpimg.imread('socialmedia.jpg')
img3 = mpimg.imread('scooter.jpg')
img4 = mpimg.imread('flower.jpeg')
```

```python
plt.figure()
plt.subplot(2,2,1)
plt.imshow(img1)

plt.subplot(2,2,2)
plt.imshow(img2)

plt.subplot(2,2,3)
plt.imshow(img3)

plt.subplot(2,2,4)
plt.imshow(img4)

plt.show()
```

## Color Bar

```python
plt.imshow(img, aspect = 1.5)
plt.colorbar()
plt.show()
```

```python
plt.imshow(img, cmap='YlOrRd')
plt.colorbar()
plt.show()
```
