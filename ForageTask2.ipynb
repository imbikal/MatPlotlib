# Data visualization

## Loading the dataset

```python
client_df = pd.read_csv('client_data.csv')
```

## Counting churn and retention

```python
churn_counts = client_df['churn'].value_counts()
```

## Visualizing with Bar Chart

```python
plt.bar(churn_counts.index, churn_counts.values, color=['skyblue', 'orange'])
plt.title('Churn Rate vs Retention')
plt.xlabel('Churn Rate (0: Retained, 1: Churned)')
plt.ylabel('Number of Clients')
plt.xticks([0, 1], ['Retention', 'Churn'])
y_ticks = range(0, max(churn_counts.values) + 1000, 1000)
plt.yticks(y_ticks)

# Add data labels with exact values
for i, value in enumerate(churn_counts.values):
    plt.text(i, value, str(value), ha='center', va='bottom', fontsize=10, color='black')

plt.show()
```
